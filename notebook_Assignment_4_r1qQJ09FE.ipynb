```python
def getListForHistogramAndBoxPlot(df,spark):
    result = spark.sql("select * from washing where temperature is not null")
    result_array = result.rdd.map(lambda  x:x.temperature).collect()
    return result_array#YOUR CODE GOES HERE
```

```python
#should return a tuple containing the two lists for timestamp and temperature
#please make sure you take only 10% of the data by sampling
#please also ensure that you sample in a way that the timestamp samples and temperature samples correspond (=> call sample on an object still containing both dimensions)
def getListsForRunChart(df,spark):
    df_s = spark.sql("select ts, temperature from washing where temperature is not null or ts is not null order by ts").sample(False, 0.1)
    df_ts = df_s.rdd.map(lambda x:x.ts).collect()
    df_temp = df_s.rdd.map(lambda x:x.temperature).collect()
    return (df_ts,df_temp)#YOUR CODE GOES HERE
```
