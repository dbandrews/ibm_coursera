```python
def skewTemperature(df,spark):    
    return spark.sql("""
SELECT 
    (
        1/(SELECT COUNT(temperature) FROM washing)
    ) *
    SUM (
        POWER(temperature-%s,3)/POWER(%s,3)
    )

as skewTemp from washing
                    """ %(meanTemperature(df,spark),sdTemperature(df,spark))).first().skewTemp
```
